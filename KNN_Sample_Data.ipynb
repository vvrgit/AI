```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn_pandas
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv("/content/KNN-Sample.csv")
data.head()
```

```python
scaler = MinMaxScaler()
features_to_scale = ["balance","income"]
scaled_values = scaler.fit_transform(data[features_to_scale])
data["norm_balance"] = scaled_values[:,0]
data["norm_income"] = scaled_values[:,1]
data.head(10)
```

```python
x11 = np.array([1000,16900])
dist_to_X11 = lambda x: np.linalg.norm(x-x11)
data["dist_toX11"] = data[["norm_balance","norm_income"]].apply(dist_to_X11,axis=1)
data.sort_values("dist_toX11")
```

```python
from sklearn.model_selection import train_test_split
default_train,default_test = train_test_split(data,test_size=0.25,random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1,metric="euclidean")
```

```python
features = ["norm_balance","norm_income"]
target = ["default"]
model.fit(default_train[features],default_train[target])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
train_predictions = model.predict(default_train[features])
test_predictions = model.predict(default_test[features])
train_conf_matrix = confusion_matrix(default_train[target],train_predictions)
pd.DataFrame(train_conf_matrix,columns=model.classes_,index=model.classes_)
```

```python
train_predictions = model.predict(default_train[features])
test_predictions = model.predict(default_test[features])
test_conf_matrix = confusion_matrix(default_test[target],test_predictions)
pd.DataFrame(test_conf_matrix,columns=model.classes_,index=model.classes_)
```

```python
train_accuracy = model.score(default_train[features],default_train[target])
test_accuracy = model.score(default_test[features],default_test[target])
print(train_accuracy,test_accuracy)
```
